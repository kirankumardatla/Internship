### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

### Solution

```python
# Getting product url links
p_url = []
i=1
while i<=3:
    p_url_data = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')))
    for j in p_url_data:
        p_url.append(j.get_attribute('href'))
        
    try:
        next_btn = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')))
        next_btn.location_once_scrolled_into_view
        next_btn.click()
        time.sleep(5)
    except:
        # stop if no next btn(less than 3 pages)
        break   
    i+=1    
len(p_url)        
```

```python
# creating Empty lists
Brand_Name=[]
Name_of_the_Product=[]
Price=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]

#scraping data from each link
for i in p_url:
    driver.get(i)
    
    # Fetching brand name
    try:
        
        Brand_Name_data = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//tr[@class="a-spacing-small po-brand"]')))
        Brand_Name.append(Brand_Name_data.text[6:])
    except :
        Brand_Name.append("-")
        
    # Fetching Title
    try:
        
        Name_of_the_Product_data = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//h1[@id="title"]')))
        Name_of_the_Product.append(Name_of_the_Product_data.text)
    except :
        Name_of_the_Product.append('-')
    # Fetching Price data
    try:
        
        Price_data = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')))
        Price.append(Price_data.text)
    except :
        Price.append('-')
        
    #Fetching Availability data
    try:
        Availability_data=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@id='availability']/span")))
        Availability.append(Availability_data.text)
    except :
        Availability.append("-")
        
    #Fetching Return or Exchange Data
    try:
        Return_Exchange_data = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//span[@class='a-declarative']/div/a")))
        Return_Exchange.append(Return_Exchange_data.text)
    except :
        Return_Exchange.append('-')
        
    #Fetching Expected Delivery Data
    try:
        Expected_Delivery_data =  WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//div[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span')))
        Expected_Delivery.append(Expected_Delivery_data.text)
    except:
        Expected_Delivery.append('-')    
    
    
    
```

```python
# Creating Data Frame
df1 = pd.DataFrame({'Name of the Product':Name_of_the_Product,'Brand Name':Brand_Name,'Price':Price,'Expected Delivery':Expected_Delivery,'Return/Exchange':Return_Exchange,'Availability':Availability,'Product Link':p_url})
df1
```

### 3. Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

### Solution

```python
#importing required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import warnings
import requests
warnings.filterwarnings('ignore')
import time
import pandas as pd
import os
```

```python
#  getting the webpage https://images.google.com/
# connecting to chrome driver
driver = webdriver.Chrome(r"C:\Users\kiran\Desktop\datatrained\jupitor notebooks\Internship\webdriver\chromedriver.exe")
# getting the link
driver.get('https://images.google.com/')
```

```python
# Create a directory to save the images
save_directory = 'scraped_images'
os.makedirs(save_directory, exist_ok=True)
```

```python
# Listing Key words
keywords = [ 'fruits', 'cars','Machine Learning', 'Guitar', 'Cakes']

#fetching image urls
for keyword in keywords:
    
    Sreach_bar=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//*[@title="Search"]')))
    Sreach_bar.clear()
    Sreach_bar.send_keys(keyword)
    Sreach_bar.send_keys(Keys.RETURN)
    fruits_data=  WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class="wXeWr islib nfEiy"]')))
    j=0
    for i in fruits_data:
        i.click()
        time.sleep(0.5)
        try:
            image_url_data=WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//img[@class="r48jcc pT0Scc iPVvYb"]'))).get_attribute('src')
        except:
            continue
        if image_url_data[:5]=='https':
            try:
                # Send a GET request to download the image
                response = requests.get(image_url_data, timeout=10)
                response.raise_for_status()

                # Save the image to the specified directory
                with open(f'{save_directory}/{keyword}_{j + 1}.jpg', 'wb') as file:
                    file.write(response.content)
                    print(f'Image {j+1}/10 of {keyword} downloaded')
                j+=1
            except:
                continue    
        else:
            continue
        if j >= 10:
            break
            
    
    
    
```

```python
#  getting the webpage https://www.flipkart.com/
# connecting to chrome driver
driver = webdriver.Chrome(r"C:\Users\kiran\Desktop\datatrained\jupitor notebooks\Internship\webdriver\chromedriver.exe")
# # getting the link
driver.get('https://www.flipkart.com/')
```

```python
#Entering input in search bar and search
search_bar = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')))
search_bar.clear()
search_bar.send_keys('Oneplus Nord')
search_bar.send_keys(Keys.RETURN)
```

```python
# Fetching urls of phones coming on 1st page
page1_urls = []
urls = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class="_1fQZEK"]')))
for url in urls:
    page1_urls.append(url.get_attribute("href"))
```

```python
len(page1_urls)
```

```python
# Creating Empty Data Frame for storing data
Smartphones = {}
Smartphones["Brand"] = []
Smartphones["Phone name"] = []
Smartphones["Colour"] = []
Smartphones["RAM"] = []
Smartphones["Storage(ROM)"] = []
Smartphones["Primary Camera"] = []
Smartphones["Display Size"] = []
Smartphones["Display Resolution"] = []
Smartphones['Secondary Camera']=[]
Smartphones["Processor"] = []
Smartphones["Processor Cores"] = []
Smartphones["Battery Capacity"] = []

Smartphones["Price"] = []
Smartphones["URL"] = []
```

```python
# Scraping data from each url of page 1
for url in page1_urls:
    driver.get(url)
    Smartphones['URL'].append(url)                                                          
    time.sleep(2)
    
    
    #Clicking on read more button
    try:
        read_more = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')))     
        read_more.click()
    except :
        print("Exception occured while moving to next page")
    
    
    #Scraping brand name of phone data
    try:
        brand_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//span[@class="B_NuCI"]')))      
        Smartphones["Brand"].append(brand_tags.text.split()[0])
    except :
        Smartphones['Brand'].append('-')
    
    
    #Scraping phone name data
    try:
        name_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][1]/table/tbody/tr[3]/td[2]/ul/li')))     
        Smartphones['Phone name'].append(name_tags.text)
    except :
        Smartphones['Phone name'].append('-')
    
    
    #Scraping phone color data
    try:
        color_tags =WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][1]/table/tbody/tr[4]/td[2]/ul/li')))      
        Smartphones['Colour'].append(color_tags.text)
    except :
        Smartphones['Colour'].append('-')
     
    
    #Scraping RAM data
    try:
        ram_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[2]/td[2]/ul/li')))                
        Smartphones['RAM'].append(ram_tags.text)
    except :
        Smartphones['RAM'].append('-')
    
    
    #Scraping ROM data
    try:
        rom_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[1]/td[2]/ul/li')))        
        Smartphones['Storage(ROM)'].append(rom_tags.text)
    except :
        Smartphones['Storage(ROM)'].append('-')
        
        
    #Scraping Primary camera data
    try:                                                                                    
        pri_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[1]/td[2]/ul/li')))
        Smartphones['Primary Camera'].append(pri_tags.text)
    except :
        Smartphones['Primary Camera'].append('-')
        
        
    #Scraping secondary camera data
    try:                                                                                    
        sec_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[6]/td[1]')))
        if sec_tags != "Secondary Camera" : 
            if WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[5]/td[1]'))).text == "Secondary Camera":
                sec_cam = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[5]/td[2]/ul/li')))
            else :
                raise NoSuchElementException
        else :
            sec_cam = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[6]/td[2]/ul/li')))
        Smartphones['Secondary Camera'].append(sec_cam.text)
    except :
        Smartphones['Secondary Camera'].append('-')
        
        
    #Scraping Display size data 
    try:
        disp_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][2]/div')))
        if disp_tags.text != "Display Features" : raise NoSuchElementException
        disp_size = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][2]/table[1]/tbody/tr[1]/td[2]/ul/li')))  
        Smartphones['Display Size'].append(disp_size.text)
    except :
        Smartphones['Display Size'].append('-')
    
    
    #Scraping display resolution data
    try:
        dires_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][2]/div')))
        if dires_tags.text != "Display Features" : raise NoSuchElementException
        disp_reso = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][2]/table[1]/tbody/tr[2]/td[2]/ul/li')))    
        Smartphones['Display Resolution'].append(disp_reso.text.replace('$',' '))
    except :
        Smartphones['Display Resolution'].append('-')
    
    
    
    #Scraping Processor data
    try:
        pro_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[2]/td[1]')))
        if pro_tags.text != "Processor Type" : raise NoSuchElementException
        processor = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[2]/td[2]/ul/li')))   
        Smartphones['Processor'].append(processor.text)
    except :
        Smartphones['Processor'].append('-')
        
    
    
    #Scraping Processor core data    
    try:                                                                                     
        core_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[3]/td[1]')))
        if core_tags.text != "Processor Core" :
            core_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[2]/td[1]')))
            if core_tags.text != "Processor Core" : 
                raise NoSuchElementException
            else :
                cores = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[2]/td[2]/ul/li')))
        else :
            cores = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[3]/td[2]/ul/li')))
        Smartphones['Processor Cores'].append(cores.text)
    except :
        Smartphones['Processor Cores'].append('-')
    
    
    
    #Scraping battery capacity data
    try:
        if WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][7]/table/tbody/tr/td[1]'))).text == "Battery Capacity" :
            bat_cep = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_3k-BhJ"][7]/table/tbody/tr/td[2]/ul/li')))
            Smartphones['Battery Capacity'].append(bat_cep.text)
        else:
            raise NoSuchElementException
    except :
        Smartphones['Battery Capacity'].append('-')
        
        
        
    #Scraping Price data
    try:
        price_tags = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')))      
        Smartphones['Price'].append(price_tags.text)
    except:
        Smartphones['Price'].append('-')
  
```

```python
#Checking lengths of all scraped data
print(len(Smartphones["Brand"]), len(Smartphones["Phone name"]), len(Smartphones["Colour"]), len(Smartphones["RAM"]), len(Smartphones["Storage(ROM)"]), len(Smartphones["Primary Camera"]), len(Smartphones["Secondary Camera"]), len(Smartphones["Display Size"]), len(Smartphones["Display Resolution"]), len(Smartphones["Processor"]), len(Smartphones["Processor Cores"]), len(Smartphones["Battery Capacity"]), len(Smartphones["Price"]), len(Smartphones['URL']))
```

```python
#scraping the data of laptop names
laptop_name = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//h3")))
for name in laptop_name:
    Laptop_Name.append(name.text)
    
#scraping the data of operating system
try:
    op_sys = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//div[@class='product-detail']/div/ul/li[1]/div/div")))
    for os in op_sys:
        Operating_sys.append(os.text)
except :
    pass


#scraping data of display of the Laptop
try:
    display = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//div[@class='product-detail']/div/ul/li[2]/div/div")))
    for disp in display:
        Display.append(disp.text)
except:
    pass


# scraping data of processor
try:
    processor = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[2]/td[3]")))
    for pro in processor:
        Processor.append(pro.text)
except :
    pass


# scraping the data of memory
try:
    memory = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[5]/td[3]")))
    for memo in memory:
        Memory.append(memo.text)
except:
    pass


# scraping data of weight
try:
    weight = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[7]/td[3]")))
    for wgt in weight:
        Weight.append(wgt.text.split('&')[1])
except:
    pass


# scraping data of dimensions
try:
    dimension = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[7]/td[3]")))
    for dim in dimension:
        Dimensions.append(dim.text.split('&')[0])
except:
    pass


# scraping data of graph processor
try:
    graph = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[6]/td[3]")))
    for gra in graph:
        Graph_proc.append(gra.text)
except:
    pass


# scraping the data of price
try:
    price = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,"//td[@class='smprice']")))
    for pri in price:
        Price.append(pri.text.replace('₹ ','Rs'))
except:
    pass
```

```python
#let's get option button from the page
opt_btn = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@role="button"]')))
opt_btn.click()
time.sleep(3)

#select billionaires from options
blns = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[1]/header/nav/div[1]/div[1]/div/div[2]/ul/li[2]/div[1]/span")))
blns.click()
time.sleep(3)

#select world billionaire
bln_list = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//li[@class="TjJgrPSg _2bNo56RE secondary"]/a')))
bln_list.click()
time.sleep(4)
```

```python
# scraping required data from the web page
# creating empty lists
Rank = []
Person_Name = []
Net_worth = []
Age = []
Citizenship = []
Source = []
Industry = []


while(True):
    
    # scraping the data of rank of the billionaires
    rank_tag = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="Table_rank___YBhk Table_dataCell__2QCve"]')))
    for rank in rank_tag:
        Rank.append(rank.text)
    
    
 
    # scraping the data  of names,ages,industry and Source of the billionaires
    name_tag = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="Table_dataCell__2QCve"]')))
    for i in range(0,len(name_tag),4):
        Person_Name.append(name_tag[i].text)
    for i in range(1,len(name_tag),4):
        Age.append(name_tag[i].text)
    for i in range(2,len(name_tag),4):
        Industry.append(name_tag[i].text)
    for i in range(3,len(name_tag),4):
        Source.append(name_tag[i].text)    
    
    
    
    # scraping the data of citizenship of the billionaires
    cit_tag = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@style="display: flex; flex: 1 0 auto; min-width: 0px;"]/div[5]')))
    for i in range(1,len(cit_tag)):
        Citizenship.append(cit_tag[i].text)
    
    
    # scraping data of net_worth of billionaires
    net_tag = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="Table_netWorth___L4R5 Table_dataCell__2QCve"]')))
    for net in net_tag:
        Net_worth.append(net.text)
    
    
    # clicking on next button
    try:
        next_button = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]/button[2]")))
        next_button.click()
    except:
        break
```

```python
print(len(Rank),
len(Person_Name),
len(Net_worth),
len(Age),
len(Citizenship),
len(Source),
len(Industry))
```

```python
# Creating Data Frame
Billionaires = pd.DataFrame({})
Billionaires['Rank'] = Rank
Billionaires['Name'] = Person_Name
Billionaires['Net Worth'] = Net_worth
Billionaires['Age'] = Age
Billionaires['Citizenship'] = Citizenship
Billionaires['Source'] = Source
Billionaires['Industry'] = Industry
Billionaires
```

### 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

### Solution

```python
#importing required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException,TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import warnings
warnings.filterwarnings('ignore')
import time
import pandas as pd
```

```python
#  getting the webpage https://www.youtube.com/
# connecting to chrome driver
driver = webdriver.Chrome(r"C:\Users\kiran\Desktop\datatrained\jupitor notebooks\Internship\webdriver\chromedriver.exe")
# # getting the link
driver.get("https://www.youtube.com/")
```

```python
# Entering RRR in search bar
time.sleep(4)
search_bar = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//input[@id="search"]')))
search_bar.clear()
search_bar.send_keys('RRR')
search_bar.send_keys(Keys.RETURN)
```

```python
#clicking a video
vid = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@id="video-title"]')))
vid[1].click()

time.sleep(10)
```

```python
# 1000 times we scroll down by 10000 in order to generate more comments
for _ in range(1000):
    driver.execute_script("window.scrollBy(0,500)")
    time.sleep(0.2)
```

```python
# creating empty lists
comments = []
Time = []
Likes = []




# scrape comments
cm = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//yt-formatted-string[@id="content-text"]')))
for i in cm:
    if i.text is None:
        comments.append("--")
    else:
        comments.append(i.text)


# scrape time when comment was posted
tm = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')))
for i in tm:
    Time.append(i.text)


# scrape the comment likes
like = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//span[@id="vote-count-middle"]')))
for i in like:
    Likes.append(i.text)
```

```python
print(len(comments),len(Time),len(Likes))
```

```python
# creating dataframe for scraped data

Youtube = pd.DataFrame({})
Youtube['Comment'] = comments[:500]
Youtube['Comment Time'] = Time[:500]
Youtube['Comment Upvotes'] = Likes[:500]
Youtube
```

### 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

### Solution

```python
#importing required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException,TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import warnings
warnings.filterwarnings('ignore')
import time
import pandas as pd
```

```python
#  getting the webpage https://www.hostelworld.com/
# connecting to chrome driver
driver = webdriver.Chrome(r"C:\Users\kiran\Desktop\datatrained\jupitor notebooks\Internship\webdriver\chromedriver.exe")
# getting the link
driver.get("https://www.hostelworld.com/")
```

```python
while True:
    try:
        # Entering 'London' in Search Bar
        Search_bar = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//input[@id="location-text-input-field"]')))
        time.sleep(2)
        Search_bar.send_keys('London')

        # clicking 1st option
        opt1=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]')))
        opt1.click()

        # Clicking Let's go Option
        go_opt=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//button[@id="search-button"]')))
        go_opt.click()
        break

    except TimeoutException:
        # In Case Website Loads Differently
        driver.close()
        driver = webdriver.Chrome(r"C:\Users\kiran\Desktop\datatrained\jupitor notebooks\Internship\webdriver\chromedriver.exe")
        driver.get("https://www.hostelworld.com/")
    
    
        
```

```python
# creating empty list & find required data
hostel_name = []
url = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
```

```python
while True:
    try:
        
        while True:    
            # Scraping Hostel Name and url
            hostel_name_data= WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//h2[@class="title title-6"]/a')))
            for i in hostel_name_data:
                if i.text != None:
                    hostel_name.append(i.text)
                else:
                    hostel_name.append("-")
                if i.get_attribute('href') != None:
                    url.append(i.get_attribute('href'))
                else:
                    url.append("-") 

            #Scraping Distance
            distance_data= WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//span[@class="description"]/span')))
            for i in distance_data:
                if i.text != None:
                    distance.append(i.text[2:])
                else:
                    distance.append("-")

            #Scraping Pvt Price
            pvt_prices_data= WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class="prices"]/div[1]')))
            for i in pvt_prices_data:
                if i.text.split('\n')[0] == 'No Privates Available':
                    pvt_prices.append(i.text.split('\n')[0])
                elif i.text.split('\n')[0] == 'Privates From':
                    try:
                        pvt_prices.append(i.text.split('\n')[1].split(' ')[1])
                    except IndexError :
                        pvt_prices.append(i.text.split('\n')[1].split(' ')[0])

            #scraping Dorms price
            dorms_price_data= WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class="prices"]/div[2]')))

            for i in dorms_price_data:
                if i.text.split('\n')[0] == 'No Dorms Available':
                    dorms_price.append(i.text.split('\n')[0])
                elif i.text.split('\n')[0] == 'Dorms From':
                    try:
                        dorms_price.append(i.text.split('\n')[1].split(' ')[1])
                    except IndexError :
                        dorms_price.append(i.text.split('\n')[1].split(' ')[0])

            # Scraping rating,reviews and over all data
            rating_data= WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="rating rating-summary-container big"]')))
            for i in rating_data:
                rating.append(i.text.split('\n')[0])
                try:
                    reviews.append(i.text.split('\n')[2][:-13])
                    over_all.append(i.text.split('\n')[1])
                except IndexError:
                    reviews.append(i.text.split('\n')[1])
                    over_all.append('No Data')
            # scraping facilities data
            facilities_data= WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="facilities-label facilities"]')))
            for i in facilities_data:
                facilities.append(i.text.replace('\n',' ,'))
            # clicking next page button    
            try:
                nxt_btn = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="pagination-item pagination-next"]')))

                nxt_btn.location_once_scrolled_into_view
                time.sleep(2)


                nxt_btn.click()
            except:
                break
        break
    except TimeoutException:
        # In Case Website Loads Differently
        while True:
            try:
                # Entering 'London' in Search Bar
                Search_bar = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//input[@id="location-text-input-field"]')))
                time.sleep(2)
                Search_bar.send_keys('London')

                # clicking 1st option
                opt1=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]')))
                opt1.click()

                # Clicking Let's go Option
                go_opt=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//button[@id="search-button"]')))
                go_opt.click()
                break

            except TimeoutException:
                # In Case Website Loads Differently
                driver.close()
                driver = webdriver.Chrome(r"C:\Users\kiran\Desktop\datatrained\jupitor notebooks\Internship\webdriver\chromedriver.exe")
                driver.get("https://www.hostelworld.com/")
```

```python
print(len(hostel_name),
len(distance),
len(pvt_prices),
len(dorms_price),
len(rating),
len(reviews),
len(over_all),
len(facilities),
len(url))
```
